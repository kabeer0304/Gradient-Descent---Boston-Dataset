```python
import numpy as np
import pandas as pd
```

```python
train_data = np.genfromtxt('Combined Training Data.csv',delimiter=',')
test_data  = np.genfromtxt('Combined Testing Data.csv',delimiter=',')  
```

```python
from sklearn import preprocessing
```

```python
t = preprocessing.minmax_scale(train_data)
t.shape
```

```python
def step_gradient(points,l_rate,M):
    N = points.shape[0]
    cols = points.shape[1]
    new_M = np.zeros(cols)
    for i in range(N):
        x = points[i,0:cols-1]
        x = np.append(x,1)
        y = points[i,cols-1]
        for j in range(cols):
            new_M[j] += (-2/N) * (y - (M*x).sum()) * x[j]
        M = M - (l_rate * new_M)
    return M
```

```python
def gd(points,l_rate,num_iter):
    cols = points.shape[1]
    M = np.zeros(cols)
    for i in range(num_iter):
        M = step_gradient(points,l_rate,M)
        print(i,'Cost: ',cost(points,M))
    return M
```

```python
def cost(points,M):
    total_cost = 0
    N = points.shape[0]
    cols = points.shape[1]
    for i in range(N):
        x = points[i,0:cols-1]
        x = np.append(x,1)
        y = points[i,cols-1]
        total_cost += (1/N) * (y - (M*x).sum()) ** 2
    return total_cost
```

```python
def run():
    l_rate = 0.00001
    num_iter = 1000
    m = gd(t,l_rate,num_iter)
    return m[:4],m[4]
```

```python
m,c = run()
print(m,c)
```

```python
def predict(x,m,c):
    N = x.shape[0]
    y = np.zeros(N)
    for i in range(N):
        X = x[i,:]
        y[i] = (m*x).sum() + c
    return y
x_test = test_data
y_pred = predict(x_test,m,c)
y_pred
```

```python
np.savetxt('Combined Cycle Power Plant',y_pred,delimiter=',')
```
